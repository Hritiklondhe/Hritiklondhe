```python
import matplotlib.pyplot as plt
```

```python
#PIE Chart
hours = [8,10,2,4]
work = ['sleep','college','food','leisture']
plt.pie(hours,labels = work,colors= ('r','b','gray','orange'),startangle = 90,autopct = '%1.1f%%')#colors= ('r','b','gray','orange')
plt.show
```

```python
#SCATTER PLOT

a = [1,2,3,4,5,6,7,8]
b = [5,4,6,2,8,7,3,9]
plt.scatter(a,b,marker = "*",s = 100)
#plt.scatter(a,b,marker,size,transprent)
```

```python
plt.scatter(b,a,marker = "*",s = 100,alpha = 0.6)
```

```python
import numpy as np
```

```python
x = np.random.randint(0,50,50)
y = np.random.randint(0,50,50)
plt.scatter(x,y)
```

```python
plt.scatter(x,y,s = x) #leftside while be smaller as compare to right side.
```

```python
plt.scatter(x,y,s = y) #bottom while be smaller as compare to top.
```

```python
plt.scatter(y,x,s = x,c= y,alpha = 0.5) #combine
```

```python
plt.scatter(y,x,s = x,c= y,alpha = 0.8,cmap='hsv')
```

```python
plt.scatter(x,y,s = y,c =x,marker = ">",alpha = 0.8,cmap = 'hsv')
```
